### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
import pandas as pd
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
```

```python
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
# Clean the 'item_price' column only if it's still an object (string) type
if chipo['item_price'].dtype == 'object':
    chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)

# Filter for products costing more than $10.00
products_over_10 = chipo[chipo['item_price'] > 10.00]
```

```python
display(products_over_10)
```

### Step 5. What is the price of each item?
###### print a data frame with only three columns item_name choice_description and product_price

```python
item_details = chipo[['item_name', 'choice_description', 'item_price']]
display(item_details)
```

### Step 6. Sort by the name of the item

```python
chipo_sorted = chipo.sort_values(by='item_name')
display(chipo_sorted.head())
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
max_price = chipo['item_price'].max()
most_expensive_items = chipo[chipo['item_price'] == max_price]
total_quantity_most_expensive = most_expensive_items['quantity'].sum()

display(f"The most expensive item(s) cost: ${max_price:.2f}")
display(f"The total quantity of the most expensive item(s) ordered is: {total_quantity_most_expensive}")
display("Details of the most expensive item(s) ordered:")
display(most_expensive_items[['item_name', 'choice_description', 'quantity', 'item_price']])
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowl_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
count_veggie_salad_bowl = veggie_salad_bowl_orders.shape[0]

display(f"'Veggie Salad Bowl' was ordered {count_veggie_salad_bowl} times.")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_multiple_orders = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
count_canned_soda_multiple = canned_soda_multiple_orders.shape[0]

display(f"'Canned Soda' was ordered more than once {count_canned_soda_multiple} times.")
```
